제공되는 Baseline Code는 본 인증 평가의 데이터 분석 방법 이해를 돕기 위한 참고용이며,

자율적으로 분석/모델링 방법을 활용하여 평가에 참여하실 수 있습니다.

## 1. 데이터 불러오기

```python
import pandas as pd

train = pd.read_csv('1. train.csv', encoding='utf-8')
test = pd.read_csv('2. test.csv', encoding = 'utf-8')
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

## 2. 데이터 전처리
예측 목표인 '초전도체 구성 요소(원소) 개수'를 y로 설정하고, 나머지 변수들을 x로 설정합니다.

다만 x를 구성하는 경우에는 예측해야하는 라벨과 모델이 일반적으로 학습할 수 없는 변수들은 제외합니다.


```python
train = train.dropna()
target = 'EC'

train_x = train.drop(target, axis=1)
train_y = train[target]

numeric_cols = train_x.select_dtypes(include=['int64', 'float64']).columns
train_x = train_x[numeric_cols]
```

## 3. 분석 모델 설계
분석을 위한 모델을 준비합니다.

여기서는 가장 간단한 LogisticRegression 모델을 사용하여 회귀 모델을 생성합니다.

```python
!pip install scikit-learn
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(random_state=42, max_iter=1000)
```

## 4. 모델 학습
분석 모델을 학습시킵니다.

```python
lr_model.fit(train_x, train_y)
```

## 5. 예측값 생성
학습한 모델을 사용하여 예측값을 생성합니다.


```python
test = test.fillna(0)
```

```python
test_x = test[numeric_cols]
```

```python
prediction = lr_model.predict(test_x)
prediction
```

## 6. 제출 파일 생성
submission 파일을 만들어서 제출합니다.

```python
submission = pd.read_csv('3. Sample_submission.csv', encoding = 'utf-8')
```

```python
submission[target] = prediction
```

```python
submission.to_csv('baseline_submit.csv', index=False)
```
